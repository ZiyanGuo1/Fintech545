```python
import numpy as np
import time

```

```python
# prepare data, similar with code in class; reimplement by python
n = 500
data = np.ones((n,n))
data = data - 0.1
for i in range(n):
    data[i,i] = 1
data[0,1]=0.735
data[1,0]=0.735
```

```python
# check data is non-psd
def is_non_psd(data):
    eigen_vals = np.linalg.eigvals(data)
    non_psd = False
    for val in eigen_vals:
        if val <0:
            non_psd = True
            break
    if not non_psd:
        print("psd")
    else:
        print("non-psd")
```

```python
# approach 1: using near_psd
def nearPSD(mat,eps=0):
    n = mat.shape[0]
    eigval, eigvec = np.linalg.eig(mat)
    val = np.matrix(np.maximum(eigval,eps))
    vec = np.matrix(eigvec)
    T = 1/(np.multiply(vec,vec) * val.T)
    T = np.matrix(np.sqrt(np.diag(np.array(T).reshape((n)) )))
    M = T * vec * np.diag(np.array(np.sqrt(val)).reshape((n)))
    return (M*M.T)
```

```python
start_time = time.time()
near_psd_1 = nearPSD(data)
end_time = time.time()
print("time consuming: {:.2f}s".format(end_time - start_time))
```
